```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\IT\Desktop\Data Analysis\AnalystLab Africa Internship\OnlineRetail.csv\cleaned_online_retail.csv",low_memory=False)
df
```

```python
# Create Revenue Column
df['revenue']=df['quantity']*df['unitprice']
df['revenue']
```

```python
df.describe()
```

```python
# Top selling products
top_products=df.groupby('description')['quantity'].sum().sort_values(ascending=False).head(10)

top_products
```

```python
# Highest Revenue Country
country_rev=df.groupby('country')['revenue'].sum().sort_values(ascending=False)

country_rev.head(10)
```

```python
df['invoicedate'].dtype  
```

```python
df['invoicedate'] = pd.to_datetime(df['invoicedate'])
```

```python
df['invoicedate'].dtype 
```

```python
#Monthly sales Trend
df['month']=df['invoicedate'].dt.to_period('M')

monthly_sales=df.groupby('month')['revenue'].sum()

monthly_sales
```

```python
# Most purchased product
df['description'].value_counts().head(10)
```

```python
# Customer purchased behavior
customer_purchase=df.groupby('customerid')['revenue'].sum()

customer_purchase.describe()
```
